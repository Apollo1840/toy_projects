```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
def itembase2datatable(items):
    keys = [key for key in items[0]]
    
    item_dict = dict([(key, []) for key in keys])
    for item in items:
        for key in keys:
            item_dict[key].append(item.get(key, None))
            
    return pd.DataFrame(item_dict)

def scatter_text(x, y, text_column, data, **kwargs):
    # Create the scatter plot
    p1 = sns.scatterplot(x, y, data=data, **kwargs)
    plt.xlabel(x)
    plt.ylabel(y)


    # Add text besides each point
    for line in range(0, data.shape[0]):
        p1.text(data[x][line] + 0.01, data[y][line],
                data[text_column][line], horizontalalignment='left',
                size='medium', color='black', weight='semibold')

    # Set title and axis labels
    if kwargs.get("title", None):
        plt.title(title)
    

    return p1
```
